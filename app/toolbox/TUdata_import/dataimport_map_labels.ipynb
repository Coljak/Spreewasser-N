```python
import os
import sys
import django

os.environ["DJANGO_ALLOW_ASYNC_UNSAFE"] = "true"
```

```python
from toolbox import models
from django.contrib.gis.geos import Point, Polygon, GEOSGeometry
import pandas as pd
import os
```

```python
csv = pd.read_csv('/app/toolbox/TUdata_import/map_labels.csv', delimiter=',')
csv
```

```python
objects = []
for _, row in csv.iterrows():
    obj = models.MapLabels(
        name=row.iloc[1],   # column index 1 → "name"
        
        map=row.iloc[6],    # column index 6 → "map"
        map_value=row.iloc[3],
        default_score=row.iloc[5],    # or pick a sensible default if CSV has no value
        order_position=0    # likewise, adjust if CSV has something
    )
    objects.append(obj)

# Bulk insert into DB
models.MapLabels.objects.bulk_create(objects)
```
